```python
#How many people have heart disease and how many people doesn't have heart disease?
df.target.value_counts()
```

```python
#Bar chart
df.target.value_counts().plot(kind="bar", color=["red", "blue"])
plt.title("heart disease values")
plt.xlabel("1=Heart disease , 0=No heart disease")
plt.ylabel("Amount")
plt.show()
```

```python
df.target.value_counts().plot(kind="pie",figsize=(7,6))
plt.legend(["Disease","No disease"])
plt.show()
```

```python
# "1" is male
# "0" is female
# How many male and female are in data set
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind="bar",color=("red","blue"))
plt.xlabel("1=Male, 0=Female")
plt.ylabel("Amount")
plt.title("Male and Female ratio")
plt.show()
```

```python
df.sex.value_counts().plot(kind="pie",figsize=(7,6))
plt.legend(["Male","Female"])
plt.title("Male and Female ratio")
plt.show()
```

```python
#people of which sex having most heart disease
pd.crosstab(df.target, df.sex)
```

```python
pd.crosstab(df.target, df.sex).plot(kind="bar")
plt.xlabel("0=no heart disease 1= Heart disease")
plt.title("Heart disease frequency for sex")
plt.show()
```

```python
# most of people having which type of chest pain most?
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind="bar",color=["r","b","g","y"])
```

```python
# people of which sex has which type of chest pain most?
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind="bar")
plt.xlabel("0=Female, 1=Male")
plt.title("Types of chest pain most in sex")
plt.legend()
```

```python
# In case of male and female most of them has "0 type" chest pain and least is "3 type"
# In case of female "0 type" and "2 type" are almost same
```

```python
#people which chest pain are most pron to have heart disease
pd.crosstab(df.cp,df.target)
```

```python
pd.crosstab(df.cp,df.target).plot(kind="bar")
plt.legend(["No haert disease","Heart disese"])
```

```python
# most of people who has "0 type" of chest pain has less chance of heart disease
```
